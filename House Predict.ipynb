```python
# Import the libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Sample dataset: square footage (x) and price (y)
square_feet = np.array([500, 750, 1000, 1250, 1500]).reshape(-1, 1)
prices = np.array([150000, 200000, 250000, 300000, 350000])
```

```python
# Create the model and train it
model = LinearRegression()
model.fit(square_feet, prices)
```

```python
# Predict price of a 1750 sq ft home
pedicted_price = model.predict([[1750]])
```

```python
print(f"Predicted price for aa 1750 sq ft home: ${predicted_price[0]:,.2f}")
```
